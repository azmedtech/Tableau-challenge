## Data Preparation

#### Merging multiple csv files from the CitiBike datasource to prepare one dataset for Q1 & Q2 2024 and one dataset for Q1 & Q2 2023

```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path
```

#### Q1 & Q2 2024 data

```python
# Load csv files
Jan_Q1_24_df = pd.read_csv("2024_Data_No_May/JC-202401-citibike-tripdata.csv", dtype={5: "object", 7: "object"})
Feb_Q1_24_df = pd.read_csv("2024_Data_No_May/JC-202402-citibike-tripdata.csv", dtype={5: "object", 7: "object"})
Mar_Q1_24_df = pd.read_csv("2024_Data_No_May/JC-202403-citibike-tripdata.csv", dtype={5: "object", 7: "object"})
Apr_Q2_24_df = pd.read_csv("2024_Data_No_May/JC-202404-citibike-tripdata.csv", dtype={5: "object", 7: "object"})
Jun_Q2_24_df = pd.read_csv("2024_Data_No_May/JC-202406-citibike-tripdata.csv", dtype={5: "object", 7: "object"})
Jul_Q3_24_df = pd.read_csv("2024_Data_No_May/JC-202407-citibike-tripdata.csv", dtype={5: "object", 7: "object"})
```

```python
# Add a new column "Month" with the appropriate month's value entered
Jan_Q1_24_df['Month'] = 'January'
Feb_Q1_24_df['Month'] = 'February'
Mar_Q1_24_df['Month'] = 'March'
Apr_Q2_24_df['Month'] = 'April'
Jun_Q2_24_df['Month'] = 'June'
Jul_Q3_24_df['Month'] = 'July'
```

```python
# Add a new column "Quarter" with the appropriate quarter's value entered
Jan_Q1_24_df['Quarter'] = 1
Feb_Q1_24_df['Quarter'] = 1
Mar_Q1_24_df['Quarter'] = 1
Apr_Q2_24_df['Quarter'] = 2
Jun_Q2_24_df['Quarter'] = 2
Jul_Q3_24_df['Quarter'] = 3
```

```python
# Add a new column "Year" with the appropriate year's value entered
Jan_Q1_24_df['Year'] = 2024 
Feb_Q1_24_df['Year'] = 2024
Mar_Q1_24_df['Year'] = 2024
Apr_Q2_24_df['Year'] = 2024
Jun_Q2_24_df['Year'] = 2024
Jul_Q3_24_df['Year'] = 2024
```

```python
# Concatenate the DataFrames along the rows
combined_2024_df = pd.concat([Jan_Q1_24_df, Feb_Q1_24_df, Mar_Q1_24_df, Apr_Q2_24_df, Jun_Q2_24_df, Jul_Q3_24_df], ignore_index=True)

# View the combined dataframe
combined_2024_df
```

```python
# Drop duplicate columns, if any
combined_2024_df = combined_2024_df.loc[:, ~combined_2024_df.columns.duplicated()]

# View the combined dataframe
combined_2024_df
```

```python
# Split the 'started_at' column into 'start_date' and 'start_time'
combined_2024_df['start_date'] = pd.to_datetime(combined_2024_df['started_at'], errors='coerce').dt.date
combined_2024_df['start_time'] = pd.to_datetime(combined_2024_df['started_at'], errors='coerce').dt.time

combined_2024_df
```

```python
# Drop rows with NaT values (unconverted data)
combined_2024_df = combined_2024_df.dropna(subset=['start_date', 'start_time'])
combined_2024_df
```

```python
# Drop the original 'started_at' column
combined_2024_df.drop(columns=['started_at'], inplace=True)
combined_2024_df
```

```python
# Split the 'ended_at' column into 'end_date' and 'end_time'
combined_2024_df['end_date'] = pd.to_datetime(combined_2024_df['ended_at'], errors='coerce').dt.date
combined_2024_df['end_time'] = pd.to_datetime(combined_2024_df['ended_at'], errors='coerce').dt.time

combined_2024_df
```

```python
# Drop rows with NaT values (unconverted data)
combined_2024_df = combined_2024_df.dropna(subset=['end_date', 'end_time'])
combined_2024_df
```

```python
# Drop the original 'ended_at' column
combined_2024_df.drop(columns=['ended_at'], inplace=True)
combined_2024_df
```

```python
# Review final dataframe information
combined_2024_df.info()
```

```python
# Save the combined DataFrame to a new CSV file
combined_2024_df.to_csv("final_2024_data.csv", index=False)
```
